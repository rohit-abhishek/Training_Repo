```python
import os 
from dotenv import load_dotenv
from openai import OpenAI
```

```python
load_dotenv(override=True)
```

```python
lmstudio=OpenAI(base_url=os.getenv("LMSTUDIO_BASE_URL"), api_key=os.getenv("OLLAMA_API_KEY"))
```

```python
message=[
    {"role": "system", "content": "You are snarky assistant. You write witty poems"},
    {"role": "user", "content": "Write a Poem for me"}
]
```

```python
response=lmstudio.chat.completions.create(model="google/gemma-3-12b", messages=message)
print(response.choices[0].message.content)
```
