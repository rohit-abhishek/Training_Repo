To demonstrate your familiarity with OpenAI API, and also Ollama, build a tool that takes a technical question,  
and responds with an explanation. This is a tool that you will be able to use yourself during the course!

```python
# imports
import os 
from dotenv import load_dotenv
from openai import OpenAI 
from IPython.display import Markdown, display

load_dotenv(override=True)
api_key=os.getenv("OPENAI_API_KEY")
```

```python
# constants
MODEL_GPT = 'gpt-4o-mini'
MODEL_LLAMA = 'llama3.2'
```

```python
# here is the question; type over this to ask something new

question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
```

```python
messages=[
    {"role": "system", "content": "You are Python Developer"}, 
    {"role": "user", "content" : question}
]
```

```python
# Get gpt-4o-mini to answer, with streaming
openai=OpenAI()
response=openai.chat.completions.create(model=MODEL_GPT, messages=messages)
```

```python
display(Markdown(response.choices[0].message.content))
```

```python
# Get Llama 3.2 to answer\
llama=OpenAI(base_url=os.getenv("OLLAMA_BASE_URL"), api_key="ollama")
response=llama.chat.completions.create(model=MODEL_LLAMA, messages=messages)
display(Markdown(response.choices[0].message.content))
```
