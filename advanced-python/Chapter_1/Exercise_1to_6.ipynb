```
Q1. Give correct order when building software - corretness (what it is supposed to do), efficiency (optimized for speed and performance), and functionality (program runs)
Ans: Functionality, Correctness, Efficiency 

Q2. How assert statement be used to check for correctness?
Ans: Assert statement is a debugging aid. It provides mechanism to validate the condition as boolean expression. IF condition validate to true program continues execution without any interruption. Good for debugging, validate condition. PRimaily used for dev and test. In optimized mode i.e. running python with -0 flag; assert statements are disabled.

Q3. What is benchmark in context of optimizing software program 
Ans: Benchmark is small but representative use case that can be used to estimate the speed of the program. Benchmark can be used to compare different versions of program to see if new implementation leads to improvement in efficiency. 

Q4. How timeit magic command be used in Python to estimate the speed the piece of code 
Ans: timeit command provides basic view of how long program took to execute. Available in Jupyter notebook. When placed in front of code snippet it will run code several times and record the runnin time for each run. The output of this command will show the summary stats of recorded time, so we can estimate the average time of code we are interested in 

Q5. 3 different types of info that are recorded and returned by cProfile in context of profiling a program??
Ans: 
a. ncalls - Number of times function was called
b. tottime - total time spent by the function without taking inot the account to other functions 
c. cumtime - time spent by function include other function calls 
d. percall - time spent for single call of the function - this can be obtained by dividing the total or cummulative time by number of calls 
e. filename:lineno - filename and line number. this information will not be avaible when calling C extension modules

Q6. On high level what is role of dis module in optimization? 
Ans: Provides low level details of program created. This is important when we want to know the low level instructions that corressponds to specific statement
