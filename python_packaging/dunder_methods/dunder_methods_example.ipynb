if we have 2 strings "hello" and "world" stored in str1 and str2 
str1 + str2 will give "helloworld" because String Class in Python has a dunder method called __add__() which performs string concatenation. 


````python
str1="hello"
str2="world" 

new_str=str1+str2 
print(new_str)
````

````python
# we can call __add__ method as well to do this 
new_str=str1.__add__(str2)
print(new_str)
````

``` 
For integer there is special add method to do addition

````python
class Counter:
    def __init__(self):
        self.value=1 
    
    def count_up(self):
        self.value+=1 

    def count_down(self):
        self.value-=1 

    def __str__(self):
        """ Print human readable object """ 
        return f"Count={self.value}"
    
    def __add__(self, other):
        """ For addition """
        if isinstance(other, Counter):
            return self.value + other.value 
        raise Exception("Invalid Type")


counter_1=Counter() 
counter_2=Counter()

counter_1.count_up()
counter_2.count_down()
counter_2.count_down() 

print(counter_1, counter_2)
print(counter_1 + counter_2)
````

````python
# however if we add counter with a number then we will get atrribute error 
print(counter_1 + 2)
````

```
__str__ vs __repr__ method

````python
class Car:
    def __init__(self, make, model, year):
        self.make=make 
        self.model=model 
        self.year=year 
    
    def __str__(self):
        """ For humar readable format """
        return f"{self.make} {self.model} {self.year}"
    
    def __repr__(self):
        """ For more detailed unambigous output. Usually used in debugging """
        return f"Car(make={self.make}, model={self.model}, year={self.year})"
    
my_car=Car("Creta", "SXO", "2024")
print(str(my_car))
print(repr(my_car))
````
